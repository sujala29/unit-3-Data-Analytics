#### Fill in blanks with a brief description of each plot type:

  
- Histogram
A histogram is a graphical representation of the distribution of a set of continuous or discrete data. In a histogram, the data is divided into a set of intervals or "bins," and the frequency or count of the data points within each bin is represented by the height of a bar. The bars are positioned adjacent to one another to form the histogram. The histogram provides a visual representation of the shape of the distribution, such as whether it is symmetrical, positively or negatively skewed, or multi-modal.


- Density Plot / KDE

A density plot is a representation of the distribution of a set of continuous data. It shows the probability density of the data at different points, rather than the raw count as in a histogram. In a density plot, the data is transformed into a smoothed line or curve that approximates the underlying probability density function.

- Bar Chart

A bar chart is a type of chart or graph that represents categorical data with rectangular bars. The bars can be vertical or horizontal and their heights or lengths correspond to the values they represent. Bar charts are often used to compare the sizes of different categories, to display data changes over time, or to show proportions and percentages.

- Line Plot
A line plot is a type of chart used to visualize the trend of one or more variables over a continuous interval or time period. A line plot represents the data points as markers connected by straight lines. Line plots are commonly used to observe trends in time series data, such as changes in stock prices or weather patterns.

- Univariate scatter plot
A univariate scatter plot is a plot of a single variable against a set of continuous or discrete data points. In a univariate scatter plot, each data point is represented by a marker on the plot, with the x-axis showing the values of the single variable and the y-axis showing the frequency or count of the data points. Univariate scatter plots are used to visualize the distribution of a single variable, to identify outliers and skewness, and to understand the relationship between the variable and other variables.

- Violin Plot
A violin plot is a type of data visualization that combines elements of a box plot and a density plot. In a violin plot, the shape of the plot shows the density of the data at different values, with a wider area representing higher density and a narrower area representing lower density. The violin plot also includes a box plot inside, which shows the median, quartiles, and outliers of the data.

- Box Plot
A box plot is a type of data visualization that displays the distribution of a set of data using a box-and-whisker representation. The box in a box plot represents the interquartile range (IQR), which covers the middle 50% of the data, and the line inside the box represents the median (middle value) of the data. The "whiskers" extend from the box to the minimum and maximum values, excluding outliers. Outliers are plotted as individual points outside the whiskers.

- Pie Chart

A pie chart is a type of data visualization that displays the proportion of different categories of data as slices of a circle. The size of each slice is proportional to the quantity it represents. Pie charts are useful for showing relative proportions, but can be less effective when the number of categories is large or the differences between the sizes of the slices are small.

- Rug Plot

A rug plot is a type of data visualization that represents the distribution of a set of data as individual lines or "ticks" that correspond to the values in the data. Rug plots are often used to provide a sense of the density of the data, especially when the data is continuous, and can be useful for exploring the distribution of the data, identifying outliers, and comparing multiple sets of data.

## Part 2 - Summary Statistics

### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
Mean is a common measure of central tendency in exploratory data analysis (EDA). EDA is the process of summarizing, visualizing, and understanding a dataset by observing its main features, patterns, and relationships between variables. Mean is one of the basic statistics used in EDA to understand the distribution of a numerical variable and its central location.

In EDA, the mean is used to understand the average value of a variable, to compare the values of two or more groups, and to get an overview of the center of the distribution. The mean is sensitive to outliers, meaning that extreme values can have a significant impact on the mean of the data. For this reason, it's essential to check for outliers and understand the distribution of the data before interpreting the mean.

The mean is a measure of the central tendency of a set of data, and can be calculated as the sum of the data divided by the number of items in the data. 

- median
The median is a measure of central tendency in a dataset that represents the midpoint of the data when the values are sorted. The median is the value that separates the higher half of the data from the lower half. It is less sensitive to outliers than the mean, and can provide a more robust estimate of the center of the data.

- mode
The mode is a measure of central tendency that represents the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), two modes (bimodal), or more than two modes (multimodal).


- central tendency
Central tendency is a statistical concept that refers to the central or typical value in a dataset. It represents a single value that summarizes the entire set of data and provides an estimate of the center of the data. The central tendency is an important aspect of exploratory data analysis (EDA), as it provides a simple and easy-to-interpret summary of the data.

- distribution
A distribution is a graphical representation of the distribution of values in a dataset. It shows the frequency or probability of occurrence of each value or range of values. Distributions are important in exploratory data analysis (EDA) because they help us understand the shape, center, spread, and outliers of a dataset. Distributions can be either continuous or discrete.

- variance
Variance is a statistical measure of the spread or dispersion of a set of data. It is a measure of how far each value in the dataset is from the mean, and it provides a way to quantify the degree of variability or scatter in the data.

- standard deviation

Standard deviation is a measure of the spread or dispersion of a set of data, and is the square root of the variance. It provides a way to quantify the degree of variability or scatter in the data, and it is expressed in the same units as the data, making it easier to interpret than the variance.

#### Fill in the blanks below with brief explanations of each type of plot listed  

- **Categorical Data**
> ***Contingency Table***

A contingency table is a table that shows the distribution of two categorical variables and the relationship between them. It displays the count or frequency of observations for each combination of categories for the two variables. The table is used in exploratory data analysis to identify patterns and relationships in categorical data.
>
> ***Clustered Bar Chart***

A Clustered Bar Chart is a type of bar graph that displays multiple bars for each category, representing different variables or sub-categories within the same category. The bars are arranged side-by-side, rather than stacked on top of each other, to better compare the values for each sub-category. This type of bar chart is used to compare and contrast the values for multiple variables within a single category.


- **Mixed Data**
> ***Individual Value Plot***
>
An Individual Value Plot (also known as a Dot Plot or Dot Chart) is a simple type of data visualization that displays individual values as dots or markers along a number line. Each dot represents a single observation and its position along the number line corresponds to its value. The plot is used to display the distribution of a small to medium-sized dataset, showing the shape, center, and spread of the data. Dot plots are especially useful for identifying outliers and skewness in the data, as well as for comparing multiple datasets.


> ***Strip Plot***
>
A Strip Plot is a type of data visualization that displays individual values as dots or markers along a number line, while adding a jitter to the position of the markers to avoid overplotting. This allows the plot to display a large number of observations while still showing the distribution of the data. Strip plots are often used in exploratory data analysis to visualize the distribution of one variable, especially when the number of observations is large. They are particularly useful for identifying outliers and skewness in the data, as well as for comparing multiple datasets.

- **Quantitative Data**
> ***Scatter plots***
>
> A Scatter Plot is a type of data visualization that displays the relationship between two variables by plotting individual data points as points on a 2D graph. Each data point is represented by its x and y values, with the x-value representing one variable and the y-value representing the other variable. The plot is used to explore the relationship between the two variables, such as association, correlation, and distribution. Scatter plots are especially useful for identifying outliers, linear and non-linear relationships, and trends in the data.

#### Fill in blanks with brief description of the following terms:
  
  
- **Correlation**

Correlation is a statistical measure that represents the strength of a linear relationship between two variables. It ranges from -1 to 1, where -1 indicates a perfect negative linear relationship, 1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship.

- **Correlation Coefficient**

The correlation coefficient is a statistical measure that quantifies the strength and direction of the linear relationship between two variables. It can range from -1 to 1, where -1 indicates a perfect negative linear relationship, 1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship. The most commonly used correlation coefficient is Pearson's correlation coefficient, which measures the linear relationship between two variables by calculating the ratio of the covariance of the two variables to the product of their standard deviations. The coefficient can be used to make inferences about the relationship between the variables, but it does not imply causality.

### Step 2 - Review Hypothesis Testing & P-Value

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

#### Fill in blanks with brief description of the following terms:

- **Hypothesis**

A hypothesis is a proposed explanation for a phenomenon or prediction about the relationship between variables.

It serves as a starting point for further investigation and testing.

- **Null Hypothesis**

The null hypothesis is a statement of no relationship or no effect, and is the default assumption in many statistical tests. It is often represented as H0. The goal of hypothesis testing is to determine whether the observed data provides enough evidence to reject the null hypothesis in favor of an alternative hypothesis, which represents a claim of a relationship or effect. The null hypothesis is tested against the alternative hypothesis, and the decision to reject or fail to reject the null hypothesis is based on the results of the statistical analysis.

- **P-Value**

A p-value is a measure of the evidence against a null hypothesis in a statistical test. It represents the probability of observing a test statistic as extreme or more extreme than the one computed from the sample data, assuming that the null hypothesis is true. The smaller the p-value, the stronger the evidence against the null hypothesis. In hypothesis testing, a commonly used threshold for the p-value is 0.05, which means that a p-value less than 0.05 is considered statistically significant and provides strong evidence to reject the null hypothesis. However, a p-value should not be used as the sole basis for making a conclusion about the null hypothesis.

#### Fill in the blanks with a brief description of each plot type 
  
- pairplots

Pair plots, also known as scatter plot matrices, are a type of plot used to visualize the relationship between multiple variables in a dataset. It is a matrix of plots, where each plot in the matrix represents the relationship between two variables. The diagonal plots are histograms or density plots, while the off-diagonal plots are scatter plots. Pair plots are useful for exploring the relationships between variables, identifying trends, and detecting outliers. They can also be used to visualize the distribution of individual variables, and to compare the relationships between variables to see if they are similar or different. Pair plots are a common tool in exploratory data analysis and can help to gain a better understanding of the structure of a dataset.

- paired scatter plots
Paired scatter plots are a type of scatter plot that are used to visualize the relationship between two variables. In a paired scatter plot, each data point represents a pair of values for the two variables. The plot shows the relationship between the two variables by plotting the values of one variable on the x-axis and the values of the other variable on the y-axis. The pattern of the data points can reveal the strength and direction of the relationship between the two variables. For example, if the data points form a clear upward trend, it suggests a positive relationship between the two variables. Paired scatter plots can be used to identify outliers, trends, and patterns in the data, and can also be used to explore the relationship between two variables in more detail.

- correlation matrix
A correlation matrix is a table showing the pairwise correlations between multiple variables in a dataset. The matrix is symmetrical, with the correlation between any two variables appearing twice, once in each cell of the matrix. Each cell in the matrix contains a correlation coefficient, which measures the strength and direction of the linear relationship between the two variables. The correlation coefficients can range from -1 to 1, where -1 indicates a perfect negative linear relationship, 1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship. A correlation matrix can be used to quickly identify strong relationships between variables and to understand the relationships between multiple variables simultaneously. It is a common tool in exploratory data analysis and can help to gain a better understanding of the structure of a dataset.

- heatmap
A heatmap is a graphical representation of data where values are represented as colors.
